# **Bivariate Analysis**

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample()
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
sns.barplot(x=tips['sex'],y=tips['total_bill'],hue=tips['smoker'])
```

```python
sns.boxplot(x=tips['sex'],y=tips['total_bill'],hue=tips['smoker'])
```

```python
sns.distplot(x=tips[tips['sex']=='Male']['total_bill'],hist=False)
sns.distplot(x=tips[tips['sex']=='Female']['total_bill'],hist=False)
```

```python
temp = tips.sample(5)
temp
```

```python
sns.heatmap(pd.crosstab(temp['sex'],temp['smoker']))
```

```python
sns.clustermap(pd.crosstab(tips['sex'],tips['smoker']))
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.lineplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.clustermap(tips.pivot_table(values='total_bill',index='day',columns='time', fill_value=0))
```
